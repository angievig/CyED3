# Regular Expressions in Python

1) Check wether the given strings contain 0xA0. Display the line and a boolean result.

````python
import re

lines = ['start address: 0xA0', 'func address: 0xC0',
         'end address: 0xFF', 'func address: 0xB0']

pattern = re.compile(r'0xA0')

for line in lines:
    print(line,bool(pattern.search(line)))
````

##Difference Between `match()` and `search()`

* match() → Only matches at the beginning of the string.
* search() → Finds the first match anywhere in the string.


````python
pattern = re.compile(r'\d+')

print(pattern.match("123abc"))   # ✅ Match at start
print(pattern.search("abc123"))  # ✅ Match found later in string
````

````python
pattern = re.compile(r'<HTML>')
print(pattern.match("<HTML>"))
print(pattern.match( " <HTML>"))
print(pattern.match(" <HTML>"[1:]))
````

## Match Objects
A MatchObject is returned when a regex pattern successfully finds a match:


```
pattern = re.compile(r'\d+')  # Match one or more digits
match = pattern.search("My number is 12345")

```
Now `match` is a MatchObject

### Methods

````python
import re

pattern = re.compile(r'(\d{3})-(\d{2})-(\d{4})')  # Matches a pattern like "123-45-6789"
match = pattern.search("My SSN is 123-45-6789")

if match:
    print("Full match:", match.group())  # 123-45-6789
    print("Group 1:", match.group(1))    # 123
    print("Group 2:", match.group(2))    # 45
    print("Group 3:", match.group(3))    # 6789
    print("All groups:", match.groups())  # ('123', '45', '6789')
````

## Using findall()
The re.findall() method returns a list of all non-overlapping matches in a string.

````python

pattern = re.compile(r'\d+')  # Match one or more digits
text = "My numbers are 123, 456, and 7890."

matches = pattern.findall(text)
print(matches)  # Output: ['123', '456', '7890']
````

##Using finditer()
The re.finditer() method returns an iterator of MatchObjects, allowing more detailed inspection.

````python


pattern = re.compile(r'\d+')  # Match one or more digits
text = "My numbers are 123, 456, and 7890."

matches = pattern.finditer(text)

for match in matches:
    print(f"Match: {match.group()}, Position: {match.span()}")
````

2) Replace all occurrences of 5 with five for the given strings

````python
lines = ['They run 5 kilometers in 5 minutes',
         'They were only 5 girls',
         'We are eating 5 apples and 5 oranges']

pattern = re.compile(r'5')

result = []
for line in lines:
    result.append(pattern.sub('five',line))
print(result)
````

3) Replace only the first occurrence of 5 with five for the given strings

````python
lines = ['They run 5 kilometers in 5 minutes',
         'They were only 5 girls',
         'We are eating 5 apples and 5 oranges']

pattern = re.compile(r'5')

result = []
for line in lines:
    result.append(pattern.sub('five',line,count=1))
print(result)
````
